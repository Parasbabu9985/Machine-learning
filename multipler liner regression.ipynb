```python
x=dataset.iloc[:,:-1]
x
y=dataset["salary"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
```

```python
x.ndim
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict(x_test)
```

```python
y_pred=lr.predict(x)
```

```python
sns.scatterplot(x='cgpa',y='salary',data=dataset)
plt.plot(dataset[['cgpa']],y_pred,color='red')
plt.show
plt.legend(['org data','predicted line'])
```
