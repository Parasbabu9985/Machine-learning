```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("polynomial_classification_data.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="data1",y="data2",data=dataset,hue="output")# hue data ko diffrentiate karne ke liye
plt.legend.location=1
plt.show()
```

```python
# ye data classifiy nahee hai to cllsificatiion kairenge
```

```python
x=dataset.iloc[:,:-1]
y=dataset["output"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[0.03,	0.99]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

# ye data ko mix kar raha hai isliye ab ise polynomial features denge


```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
