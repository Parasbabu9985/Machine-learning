```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
dataset=pd.read_csv("house_prices_sample.csv")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["price"].fillna(dataset["price"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=dataset.corr(),annot=True)
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["price"]
x,y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
sc.fit(x)

x=pd.DataFrame(sc.transform(x),columns=x.columns)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
lr = LinearRegression()
```

```python
#y_train = y_train.ravel()
lr.fit(x_train, y_train)
```

```python
print(x_train.shape, y_train.shape)
```

```python
lr.score(x_test,y_test)
```

```python
#for check error in model
```

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error
```

```python
# linear regreestion
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,lr.coef_)
plt.title('linearRegration')
plt.xlabel("columns")
plt.ylabel("coef")
plt.show
```

```python
# chek error given two value(orinal0r y true, pridict value)
print(mean_squared_error(y_test,lr.predict(x_test)))
print(mean_absolute_percentage_error(y_test,lr.predict(x_test)))
```

# lasso


```python
la=Lasso(alpha=0.5)
la.fit(x_train,y_train)
la.score(x_test,y_test)
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,lr.coef_)
plt.title('Lasso')
plt.xlabel("columns")
plt.ylabel("coef")
plt.show
```

```python
# chek error
print(mean_squared_error(y_test,la.predict(x_test)))
print(mean_absolute_percentage_error(y_test,la.predict(x_test)))
```

# Ridge

```python
rg=Ridge(alpha=10)
rg.fit(x_train,y_train)
rg.score(x_test,y_test)
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,lr.coef_)
plt.title('Ridge')
plt.xlabel("columns")
plt.ylabel("coef")
plt.show
```

```python
# chek error
print(mean_squared_error(y_test,rg.predict(x_test)))
print(mean_absolute_percentage_error(y_test,rg.predict(x_test)))
```
